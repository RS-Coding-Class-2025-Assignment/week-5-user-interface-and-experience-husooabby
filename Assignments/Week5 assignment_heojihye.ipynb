```python
import tkinter as tk
from tkinter import ttk
import time
import threading


root = tk.Tk()
root.title("Progress bar")

progress = ttk.Progressbar(root, orient = "vertical", length = 300, mode = "determinate")
progress.pack()


label = tk.Label(root, text = "Hello!")
label.pack(pady = 20)

stop_thread = False
pause_thread = False   


    
def block_gui():
    global stop_thread
    for i in range(101):       
        if stop_thread:
            break

        while pause_thread and not stop_thread:
            time.sleep(0.1)
            root.update()
        
        if stop_thread:
            break

        label.config(text = f"{i}%")
        root.update()
        progress['value'] = i
        progress.step(1)
        time.sleep(0.1)


        
def start():
    global stop_thread, pause_thread 
    progress['value'] = 0
    stop_thread = False
    pause_thread = False
    t = threading.Thread(target = block_gui)
    t.start()
    reset_button.config(state = tk.NORMAL)
    block_button.config(state = tk.DISABLED)
    pause_button.config(state = tk.NORMAL)
    resume_button.config(state = tk.DISABLED)
    


def reset_fields():
    global stop_thread
    stop_thread = True
    progress['value'] = 0
    label.config(text = "Hello!")
    reset_button.config(state = tk.DISABLED)
    block_button.config(state = tk.NORMAL)
    pause_button.config(state = tk.DISABLED)
    resume_button.config(state = tk.DISABLED)


def pause():
    global pause_thread
    pause_thread = True
    block_button.config(state = tk.DISABLED)
    pause_button.config(state = tk.DISABLED)
    reset_button.config(state = tk.DISABLED)
    resume_button.config(state = tk.NORMAL)

def resume():
    global pause_thread
    pause_thread = False
    
    pause_button.config(state = tk.NORMAL)
    reset_button.config(state = tk.NORMAL)
    block_button.config(state = tk.DISABLED)
    resume_button.config(state = tk.DISABLED)


block_button = tk.Button(root, text = "Run", command= start)
block_button.pack(pady = 10)

reset_button= tk.Button(root, text="Reset", command=reset_fields)
reset_button.pack(pady = 10)
reset_button.config(state = tk.DISABLED)

pause_button = tk.Button(root, text="Pause", command= pause)    
pause_button.pack(pady = 10)
pause_button.config(state = tk.DISABLED)


resume_button = tk.Button(root, text="Resume", command=resume)
resume_button.pack(pady = 10)
resume_button.config(state = tk.DISABLED)




root.mainloop()
```
